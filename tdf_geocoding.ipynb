```python
latitudes = []
longitudes = []

for starting_point in starting_points: 
    result = arcgis.geocode(starting_point)
    if result and len(result):
        longitude = result.longitude
        latitude = result.latitude
    else:
        longitude = "N/A"  
        latitude = "N/A"
    latitudes.append(latitude) 
    longitudes.append(longitude) 
```

```python
df_stages['starting_latitudes'] = latitudes
df_stages['starting_longitudes'] = longitudes
```
